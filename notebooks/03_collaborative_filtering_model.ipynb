```python
import pandas as pd
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate

# 1. Oylama Verisini Yükleme 
# Collaborative Filtering (CF) metinlerle değil, sadece kullanıcı oylarıyla ilgilenir.
ratings_df = pd.read_csv('../csv-dosyalari/ratings.csv')
print(f"Modele Eğitilecek Toplam Puanlama Sayısı: {len(ratings_df)}")

# 2. SVD Matrisi İçin Veri Formatlama
# Surprise kütüphanesinin okuyabilmesi için puanlama ölçeğini (0.5 ile 5.0 yıldız) belirtiyoruz
reader = Reader(rating_scale=(0.5, 5.0))

# Verimizi User-Item matrisine (Kullanıcı-Film-Puan) dönüştürüyoruz
data = Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], reader)
print("--- VERİ SVD MATRİSİ İÇİN HAZIRLANDI ---\n")

# 3. SVD (Matrix Factorization) Algoritması (PROPOZAL KANITI)
# Proje önerinizde vaat ettiğiniz "Matrix Factorization (SVD)" yöntemini başlatıyoruz
svd_model = SVD(random_state=42)

# 4. K-Fold Cross Validation (Çapraz Doğrulama) ile Test
# Akademik olarak modelin gücünü kanıtlamak için veriyi 5 parçaya bölüp eğitiyor ve hata payını (RMSE) ölçüyoruz.
print("SVD Algoritması Eğitiliyor ve Hata Payı (RMSE) Hesaplanıyor...")
print("(100.000 satır işlendiği için bu işlem 10-15 saniye sürebilir, lütfen bekleyin)\n")

sonuclar = cross_validate(svd_model, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)

print("\n--- BAĞIMSIZ İŞBİRLİKÇİ FİLTRELEME (CF) BAŞARISI ---")
print(f"SVD Modelinin Ortalama RMSE (Hata Payı) Değeri: {sonuclar['test_rmse'].mean():.4f}")
```
